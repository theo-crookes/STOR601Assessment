<h3> Replicable

The definition for replicable from https://prezi.com/view/UFn6tMCwgPK4YHf3VrlH/ is being able to write new code from a conceptual description. The pseudocode is best for this because it is the easiest to understand and then write a function doing the same. 

The Python is next best because python syntax is easy to understand and the comments will help the user follow along and write their own. The C++ is next because although C++ syntax is hard to understand, the comments can still help the user follow along. The R package is the hardest because although there is the DESCRIPTION explaining what each function does, to go and see the actual functions you have to know how to enter the package and see the C++ code itself, for that reason it is the hardest to replicate. 

<h3> Re-runnable

In terms of how easy they are to re-run, I placed the R package as the easiest because it benefits from the speed of the C++ whilst being easier to re-run because R is easier to use and work with. Next I put the C++ because it is faster to re-run than Python, since C++ is in general faster to run. Next I put the Python in 3rd place and then put the pseudocode as 4th. The reason I put the pseudocode as 4th was because you can't really run it, however, if you considered re-running the pseudocode as simply re-reading it then it depends on how fast you read and how well you understood it before, however the other algorithms are fast so it is likely that they would have finished before finishing reading the pseudocode, but it is hard to rank the pseudocode here for this reason. 

<h3> Repeatable

All of the results from the Python, C++ and the R package implementations should be equally repeatable as there is no stochasticity in any of the programs, so they will all give the same output every time, so are equally as repeatable as the next. The reason that pseudocode is placed last is because the concept of repeating it is not very well defined. 

<h3> Reproducible

Reproducibility is defined as other users being able to produce the same results with the same code+data (as defined in https://prezi.com/view/UFn6tMCwgPK4YHf3VrlH/). The R package implementation and Python are both easy to use, I ranked the R package implementation slightly higher simply because R is slightly more user friendly than Python, so users are less likely to make a mistake. C++ is harder to use but still if a user knows how to use it, the results should be reproducible. Overall, all 3 are reproducible and were just ranked in terms of how user friendly each of the software is to understand, with the R package being the easiest with the addition of things like the DESCRIPTION.

I ranked the pseudocode the lowest because two people could read the same pseudocode, then develop their own code and this code be different, so results the pseudocode produces (the functions people write after reading it) are not always the same. 

<h3> Reusable

To be reusable consists of two components, being able to be used again but also to be modified, that is why I have given two rankings for usability and modificability respectively. 

With respect to how easy they are to reuse, the pseudocode is the best because it can easily be read again, then the R package is the next easiest because it can easily be installed anywhere using devtools and then can be used again. It also uses R which is simple to understand and use. The next easiest is the Python implementation, because python is more user friendly than C++ and C++ is the least easy to reuse. 

With respect to how easy they are to modify, the pseudocode is again the best because it is the easiest to understand what is happening and make an edit to, especially because you do not need to actually code something correctly, only explain what you would like to happen. The next easiest to modify is the Python because Python has easier syntax to understand than C++. The next easiest to modify is C++, it is easier to modify than the R package because to modify the R package would involve the knowledge of the structure of R packages, how to enter them and change them and then you would need to modify the C++ within the package, so it is simpler to modify the C++ version.
